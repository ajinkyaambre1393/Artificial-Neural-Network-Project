```python
###############################Data Preprocessing##############################################
```

```python
#joining both dataset
df1 = pd.DataFrame(dataset_train)
df2 = pd.DataFrame(dataset_test)
dataset_joined = [df1, df2]
dataset_joined = pd.concat(dataset_joined)
```

```python
#Filling the null values using mean
dataset_joined['Age'].fillna(dataset_joined['Age'].mean(), inplace = True)

dataset_joined['Fare'].fillna(dataset_joined['Fare'].mean(), inplace = True)
```

```python
#Filling 0 in null values because mean of survived comes out to be 0.2612681436210848
dataset_joined['Survived'].fillna(0, inplace = True)
```

```python
#Filling null values in Embarked with 'S' as more number of occurences compare to others
dataset_joined['Embarked'].value_counts()
dataset_joined['Embarked'].fillna('S', inplace = True)
```

```python
#dropping unwanted columns 
drop_cols = ['Cabin', 'Name','PassengerId', 'Ticket']
dataset = dataset_joined.drop(drop_cols, axis = 1)
dataset
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
dataset['Embarked'] = labelencoder_X_1.fit_transform(dataset['Embarked'])
labelencoder_X_2 = LabelEncoder()
dataset['Sex'] = labelencoder_X_2.fit_transform(dataset['Sex'])
```

```python
# Creating dataframe X and y
X = dataset.iloc[:, 0:7 ].values
y = dataset.iloc[:, 7].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler() 
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
###############################Artificial Neural Network##############################################
```

```python
#importing libraries and packages for keras
import keras
from keras.models import Sequential          #initialize neural network
from keras.layers import Dense               #building layers of ANN

```

```python
#initiliazing ANN
classifier = Sequential()
```

```python
#adding input layer and first hidden layer
classifier.add(Dense(4, kernel_initializer= 'uniform', activation = 'relu', input_dim = 7))

```

```python
#adding the second hidden layer
classifier.add(Dense(4, kernel_initializer= 'uniform', activation = 'relu'))
```

```python
#adding the output layer
classifier.add(Dense(1, kernel_initializer= 'uniform', activation = 'sigmoid'))
```

```python
#Compiling ANN (Stochastic gradient)
classifier.compile(optimizer = 'adam',loss='binary_crossentropy', metrics = ['accuracy'] )
```

```python
#fitting the ANN to the training set
classifier.fit(X_train, y_train,batch_size = 10, epochs = 100)
```

```python
#predicting the test dataset results. We are cosidering that y_pred > 0.5 is survival 
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
```
